##先從MySQL中抓出company的資料，並寫入檔案中

```python
import MySQLdb 
conn = MySQLdb.connect(host='10.120.31.4',user='threela',passwd='123456',charset='utf8')
curs = conn.cursor()
conn.select_db('threela')
curs.execute('SELECT * FROM threela.company;')
result = curs.fetchall()
try:
   
    with open('Map/companyList.txt','a') as companyList:
        for recrod in result:
            #print record[0],record[2]
            companyId = recrod[0].encode('utf-8')
            companyName = recrod[2].encode('utf-8')
            companyList.write('%s,%s \n'%(companyId,companyName))
    companyList.close()

finally:
    curs.close()
    conn.close()        
```
